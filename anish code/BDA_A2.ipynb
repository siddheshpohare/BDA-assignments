```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# --------------------------
# 1. Load the dataset
# --------------------------
# Example: Load your dataset (replace 'your_file.csv' with the actual file)
df = pd.read_csv("segmentation data.csv")

# For demonstration, let's simulate some data similar to your schema
# import numpy as np

# np.random.seed(42)
# n_samples = 1000
# df = pd.DataFrame({
#     'Sex': np.random.choice([0, 1], size=n_samples),
#     'Marital status': np.random.choice([0, 1], size=n_samples),
#     'Age': np.random.randint(18, 77, size=n_samples),
#     'Education': np.random.choice([0, 1, 2, 3], size=n_samples),
#     'Income': np.random.randint(35832, 309364, size=n_samples),
#     'Occupation': np.random.choice([0, 1, 2], size=n_samples),
#     'Settlement size': np.random.choice([0, 1, 2], size=n_samples),
# })

# --------------------------
# 2. Preprocess the data
# --------------------------
features = df[['Sex', 'Marital status', 'Age', 'Education', 'Income', 'Occupation', 'Settlement size']]
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# --------------------------
# 3. Apply K-Means Clustering
# --------------------------
k = 4  # You can change the number of clusters
kmeans = KMeans(n_clusters=k, random_state=42)
df['Cluster'] = kmeans.fit_predict(scaled_features)

# --------------------------
# 4. Visualize Clusters
# --------------------------
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Age', y='Income', hue='Cluster', palette='Set2')
plt.title('Clusters based on Age and Income')
plt.xlabel('Age')
plt.ylabel('Income')
plt.grid(True)
plt.show()
```

Viva-Ready Explanation
 What is Cluster Analysis?

Cluster analysis is an unsupervised learning technique that groups similar data points into clusters. It helps in pattern discovery without prior labeling.
 Why K-Means?

    K-Means is simple, scalable, and efficient for large datasets.

    It minimizes the distance between data points and their cluster centroid.
