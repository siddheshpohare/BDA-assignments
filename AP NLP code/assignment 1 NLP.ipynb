```python
import pandas as pd
from langdetect import detect
from nltk.tokenize import word_tokenize, sent_tokenize
import nltk
nltk.download('punkt_tab')
```

```python
df = pd.read_csv("NewsCategorizer.csv")

df.head()
```

```python
def detect_language(text):
    try:
        return detect(text)
    except:
        return "unknown"
df["language"] = df["headline"].apply(detect_language)
df.head()   
```

```python
df["language"].value_counts()
```

```python
df["word_tokens"] = df["headline"].apply(word_tokenize)
df["sentence_tokens"] = df["headline"].apply(sent_tokenize)
df.head()
```

```python
df["word_tokens"] = df["word_tokens"].apply(len)
df["sentence_tokens"] = df["sentence_tokens"].apply(len)
df.head()
```

```python
df.to_csv("News Category Dataset Updated.csv", index = False)

```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
sns.countplot(
    y='language',
    data=df,
    order=df['language'].value_counts().index
)

plt.title("Language Distribution")
plt.show()
```

```python
sns.histplot(
    df['word_tokens'],
    bins=30,
    kde=True
)

plt.title("Word Count Distribution")
plt.show()
```
